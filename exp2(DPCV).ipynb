```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
img=cv2.imread('/content/gdrive/MyDrive/vertical.png',0)
img2=cv2.imread('/content/gdrive/MyDrive/vertical_high.jpg',0)
```

```python
f=np.fft.fft2(img)
fshift=np.fft.fftshift(f) #shifting DC component to centre which is previously at corners
magnitude_spectrum=20*np.log(np.abs(fshift))
```

```python
plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
def magnitude_spectrum(img):
  f=np.fft.fft2(img)
  fshift=np.fft.fftshift(f) #shifting DC component to centre which is previously at corners
  magnitude_spectrum=20*np.log(np.abs(fshift))
  plt.subplot(121),plt.imshow(img, cmap = 'gray')
  plt.title('Input Image'), plt.xticks([]), plt.yticks([])
  plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
  plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
  plt.show()

```

```python
magnitude_spectrum(img)
```

```python
magnitude_spectrum(img2)
```
